```python
target_columns = ['Bat First', 'Bat Second', 'Winner']
values_to_be_dropped =['Pune Warriors','Rising Pune Supergiant','Rising Pune Supergiants','Kochi Tuskers Kerala']

data = data[~data[target_columns].isin(values_to_be_dropped).any(axis=1)]
```

```python
# Standardizing Venue Names
venue_map = {
    "M.Chinnaswamy Stadium": "M Chinnaswamy Stadium",
    "Sardar Patel Stadium": "Narendra Modi Stadium",
    "Punjab Cricket Association IS Bindra Stadium": "Punjab Cricket Association Stadium",
    "Feroz Shah Kotla": "Arun Jaitley Stadium",
    "Zayed Cricket Stadium": "Sheikh Zayed Stadium",
    "De Beers Diamond Oval": "Kimberley Oval",
    "Rajiv Gandhi International Stadium": "Uppal Stadium",
    "Dr Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium": "ACA-VDCA Stadium",
    "Shaheed Veer Narayan Singh International Stadium": "Raipur International Cricket Stadium",
    "JSCA International Stadium Complex": "Ranchi Stadium",
    "Barsapara Cricket Stadium": "Assam Cricket Association Stadium"
}
data["Venue"] = data["Venue"].replace(venue_map)
```

```python
data['Batter'].unique()
```

```python
# Standardizing Player Names
data["Batter"] = data["Batter"].str.strip().str.title()
data["Non Striker"] = data["Non Striker"].str.strip().str.title()
data["Bowler"] = data["Bowler"].str.strip().str.title()
data["Player Out"] = data["Player Out"].str.strip().str.title()
```

```python
data.dtypes #to display data-type of each column
```

```python
# Convert Ball Rebowled to boolean
data['Ball Rebowled'] = data['Ball Rebowled'].astype(bool)
# Convert Runs to Get to integer
data['Runs to Get'] = data['Runs to Get'].astype(int)
# Convert Player Out Runs to integer
data['Player Out Runs'] = data['Player Out Runs'].astype(int)
# Convert Player Out Balls Faced to integer
data['Player Out Balls Faced'] = data['Player Out Balls Faced'].astype(int)
```

```python
categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
print(categorical_columns)
```

```python
data['is_six'] = (data['Batter Runs'] == 6).astype(int)
```

```python
data['Method'].unique()
```

```python
valid_methods = ['bowled', 'caught', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
data['bowler_wickets'] = data['Method'].isin(valid_methods).astype(int)
```

```python
data['is_dot_ball'] = (
                      (data['Runs From Ball'] == 0)
                      & (data['Extra Runs'] == 0)
                      & (data['Valid Ball'] == 1)
                    ).astype(int)
```
